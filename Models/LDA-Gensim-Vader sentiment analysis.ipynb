```python
import pandas as pd

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import re
import string
analyzer = SentimentIntensityAnalyzer()
def sentiment_analyzer_scores(review, nth):
    score_list = []
    for i in range(nth):
        score = analyzer.polarity_scores(review[i])
        for cat, point in score.items():
            if cat == 'compound':
                score_list.append(point)
    return score_list
        
    
        
        

```

```python
top_10k = pd.read_csv('/Users/jcheah/Documents/MITB/Term 2/Text analytics/Project/top10k_en.csv')
reviews = top_10k['text'] 
reviews_score = sentiment_analyzer_scores(reviews, len(reviews))
reviews_score = pd.DataFrame(reviews_score)

```

```python
top_10k['reviews_score'] = reviews_score

top_10k
```

```python
import gensim
import pandas as pd
from nltk.tokenize import word_tokenize

import pyLDAvis.gensim

df = pd.read_csv('/Users/jcheah/Documents/MITB/Term 2/Text analytics/Project/top10k_en.csv')
docs=df['text'].apply(word_tokenize)
d= gensim.corpora.Dictionary(docs)
#convert docs to vec
v = [d.doc2bow(doc) for doc in docs]
lda = gensim.models.ldamodel.LdaModel(corpus=v, id2word=d, num_topics=10)
m = lda[v[0:len(v)]]


import operator 
y = []
for x in m:
    a = max(x, key = operator.itemgetter(1))
    y.append(a[0])
y
```

```python
top_10k['topic'] = list(y)
top_10k
topic_dict = {0:'Food', 1:'Food', 2:'Service',
             3:'Food', 4:'food', 5:'Food',
             6:'Food', 7:'Flavor', 8:'Ambience', 9:'Food'}

top_10k = top_10k.replace({'topic':topic_dict})

top_10k
```

```python
top_10k_score = top_10k.loc[:, ['name', 'topic', 'reviews_score']]

round(abs((top_10k_score.groupby(['name', 'topic']).sum() / top_10k_score.groupby(['name', 'topic']).count())*5), 1)
```
